<span style='long;size'>_Manejo de datos faltantes_</span>

```python
df.notnull()
```

```python
df.notnull().apply(pd.Series.value_counts)
```

Estos resultados pueden ser interpretados de la siguiente manera:
* Si en _False_ dice "NaN" es porque en dicha columna no se encontraron datos faltantes.
* El _apply_ viene a realizar la misma función de un _for_ para cada columna. 

```python
available = ((df.notnull()) & (df != 0))
```

```python
available
```

```python
available.all(axis=1).value_counts()
```

```python
mask = available['worldwide_gross']
```

```python
df = df[mask]
df
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
df = df.drop('movie_title', axis=1)
df
```

```python
df.columns
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
df = df.drop('duration.1', axis=1)
df
```

```python
df1 = df1.drop('movie_name', axis=1)
df1
```

```python
df1 = df1.drop('release_date', axis=1)
df1
```

```python
values = imputer.fit_transform(df1)
values
```
